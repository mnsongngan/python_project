### EDA

```python
# Create figure
fig, axs = plt.subplots(2, 4, figsize = (20, 10))
axs = axs.flatten()
# Draw bar plots
for i, col in enumerate(['estimated_age', 'age_range', 'channel', 'coupon', 'clicked', 'last_step', 'nb_units', 'order_value']):
    if col in ['estimated_age', 'order_value']:
        sns.histplot(df[col], ax=axs[i], kde=False, bins=28)  # Histogram for 'order_value' and 'estimated_age'
        axs[i].set_title(col)
    else:
        value_counts = df[col].value_counts(normalize = True).reset_index()
        value_counts.columns = [col, 'proportion']  # Rename columns
        value_counts['proportion'] = value_counts['proportion'] * 100
        sns.barplot(ax = axs[i], y = 'proportion', x = col, data = value_counts, palette = 'Blues_r',order = sorted(value_counts[col].unique()))
        axs[i].set_title(col)
        # Add data labels
        for p in axs[i].patches:
            axs[i].annotate(format(p.get_height(), '.0f'),
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha = 'center', va = 'center',
                        xytext = (0, 10),
                        textcoords = 'offset points')

plt.tight_layout()
```

```python
# Click-through rate:
cth = df.groupby(['channel'])['clicked'].mean()
print(cth)
```

```python
# Click-through rate: By Channel, By age-range, By cou
fig, axs = plt.subplots(1,2, figsize = (10, 4))
axs = axs.flatten()
# Draw bar plots
for i, col in enumerate(['age_range', 'coupon']):
        ctr = df.groupby([col, 'channel'])['clicked'].mean().reset_index()
        ctr.columns = [col, 'channel', 'CTR']
        tctr = ctr.pivot(index=col, columns='channel', values='CTR').fillna(0)
        # Plot
        ax = axs[i]
        tctr.plot(kind='bar', stacked=False, ax=ax, colormap='tab20b')
        ax.set_title(f'Click-Through Rate by {col.capitalize()} and Channel')
        ax.set_xlabel(col.capitalize())
        ax.set_ylabel('Click-Through Rate (CTR)')
        ax.legend(title='Channel')
    
    # Add data labels
        for p in ax.patches:
            height = p.get_height()
            if height > 0:
                ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height ),
                        ha='center', va='bottom', fontsize=8, color='black')
plt.tight_layout()
plt.show()
```

Click-through rate (CTR) is different between groups:
* SMS shows a higher CTR than Email
* Except for customers in the range of 31-45, the CTR of email is higher.
   
Overall, the CTR is not bad. However, according to Layerise (2023), "the click-through rate of SMS marketing is typically higher than email marketing. SMS marketing typically has a click-through rate of around 19%, while email marketing typically has a click-through rate of around 3%." 
### => Here, SMS underperformed and email overperformed.

### Conversion rate

```python
# Lead Conversion rate - assume whose last_step is 'added to cart' 'purchased' 'saw review' 'payment page' are seen as leads
lead_steps = ['added to cart', 'purchased', 'saw review', 'payment page']
df['leads'] = df['last_step'].apply(lambda x: 1 if x in lead_steps else 0)
# lead conversion rate (lcr): By Channel, By age-range, By cou
fig, axs = plt.subplots(1,2, figsize = (12,5))
axs = axs.flatten()
# Draw bar plots
for i, col in enumerate(['age_range', 'coupon']):
        ctr = df.groupby([col, 'channel'])['leads'].mean().reset_index()
        ctr.columns = [col, 'channel', 'LCR']
        tctr = ctr.pivot(index=col, columns='channel', values='LCR').fillna(0)
        # Plot
        ax = axs[i]
        tctr.plot(kind='bar', stacked=False, ax=ax, colormap='tab20b')
        ax.set_title(f'Lead Conversion Rate by {col.capitalize()} and Channel')
        ax.set_xlabel(col.capitalize())
        ax.set_ylabel('Lead Conversion Rate (LCR)')
        ax.legend(title='Channel')
    
    # Add data labels
        for p in ax.patches:
            height = p.get_height()
            if height > 0:
                ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height + 0.002),
                        ha='center', va='bottom', fontsize=8, color='black')
plt.tight_layout()
plt.show()
```

```python
# Customer Conversion rate - as whose last_step is 'purchased' 
df['cus'] = df['last_step'].apply(lambda x: 1 if x == 'purchased' else 0)
# customer conversion rate (ccr): By Channel, By age-range, By cou
fig, axs = plt.subplots (1,2, figsize = (12,5))
axs = axs.flatten()
# Draw bar plots
for i, col in enumerate(['age_range', 'coupon']):
        ctr = df.groupby([col, 'channel'])['cus'].mean().reset_index()
        ctr.columns = [col, 'channel', 'CCR']
        tctr = ctr.pivot(index=col, columns='channel', values='CCR').fillna(0)
        # Plot
        ax = axs[i]
        tctr.plot(kind='bar', stacked=False, ax=ax, colormap='tab20b')
        ax.set_title(f'Customer Conversion Rate by {col.capitalize()} and Channel')
        ax.set_xlabel(col.capitalize())
        ax.set_ylabel('Customer Conversion Rate (CCR)')
        ax.legend(title='Channel')
        

    # Add data labels
        for p in ax.patches:
            height = p.get_height()
            if height > 0:
                ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', fontsize=8, color='black')

plt.tight_layout()
plt.show()
```

* Can reduce coupon for 46-60 to maximize profit

```python
average_order_value = df.groupby(['age_range', 'coupon', 'channel'])['order_value'].mean().reset_index()

# Calculate ROI ratio
average_order_value['ROI'] = average_order_value.apply(
    lambda row: row['order_value'] / (0.05 if row['channel'] == 'SMS' else 0.075),
    axis=1
)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs = axs.flatten()

# Draw bar plots for 'age_range' and 'coupon' grouped by 'channel' with ROI value
for i, col in enumerate(['age_range', 'coupon']):
    roi_data = average_order_value.groupby([col, 'channel'])['ROI'].mean().reset_index()
    roi_pivot = roi_data.pivot(index=col, columns='channel', values='ROI').fillna(0)

    # Plot
    ax = axs[i]
    roi_pivot.plot(kind='bar', stacked=False, ax=ax, colormap='tab20b')
    ax.set_title(f'Return on Investment (ROI) by {col.capitalize()} and Channel')
    ax.set_xlabel(col.capitalize())
    ax.set_ylabel('Return on Investment (ROI)')
    ax.legend(title='Channel')

    # Add data labels
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', fontsize=8, color='black')

plt.tight_layout()
plt.show()
```
